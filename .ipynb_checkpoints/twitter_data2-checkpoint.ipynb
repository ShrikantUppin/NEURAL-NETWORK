```python
import re

#Clean text from noise
def clean_text(text):
    
    # URL remover...
    text = re.sub(r'http\S+', '', text)
    
    #Filter to allow only alphabets
    text = re.sub(r'[^a-zA-Z\']', ' ', text)
    
    #Remove Unicode characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # removing single character word..
    text = re.sub(r"\s+[a-zA-Z]\s+", " ", text)
    
    # remove words between 1 and 2
    shortword = re.compile(r'\W*\b\w{1,2}\b')
    text = re.sub(shortword,'', text)
    
    #Convert to lowercase to maintain consistency
    text = text.lower()
       
    return text
```

```python
data_pre['clean_text'] = data_pre.text.apply(lambda x: clean_text(x))
```

```python
for index, text in enumerate(data_pre['clean_text'][10:15]):
    print(index+1,".",text)
```

```python
data_pre.head()
```

```python
#Generate word frequency
def gen_freq(text):
    
    from wordcloud import STOPWORDS
    
    STOP_WORDS = STOPWORDS
    
    #Will store the list of words
    word_list = []

    #Loop over all the tweets and extract words into word_list
    for tw_words in text.split():
        word_list.extend(tw_words)

    #Create word frequencies using word_list
    word_freq = pd.Series(word_list).value_counts()
    
    #Drop the stopwords during the frequency calculation
    word_freq = word_freq.drop(STOP_WORDS, errors='ignore')
    
    return word_freq

#Check whether a negation term is present in the text
def any_neg(words):
    for word in words:
        if word in ['n', 'no', 'non', 'not'] or re.search(r"\wn't", word):
            return 1
    else:
        return 0

#Check whether one of the 100 rare words is present in the text
def any_rare(words, rare_100):
    for word in words:
        if word in rare_100:
            return 1
    else:
        return 0

#Check whether prompt words are present
def is_question(words):
    for word in words:
        if word in ['when', 'what', 'how', 'why', 'who']:
            return 1
    else:
        return 0
```

```python

word_freq = gen_freq(data_pre.clean_text.str)

#100 most rare words in the dataset
rare_100 = word_freq[-100:]

#Number of words in a tweet
data_pre['word_count'] = data_pre.clean_text.str.split().apply(lambda x: len(x))

#Negation present or not
data_pre['any_neg'] = data_pre.clean_text.str.split().apply(lambda x: any_neg(x))

#Prompt present or not
data_pre['is_question'] = data_pre.clean_text.str.split().apply(lambda x: is_question(x))

#Any of the most 100 rare words present or not
data_pre['any_rare'] = data_pre.clean_text.str.split().apply(lambda x: any_rare(x, rare_100))

#Character count of the tweet
data_pre['char_count'] = data_pre.clean_text.apply(lambda x: len(x))
```
